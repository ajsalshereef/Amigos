```
import pandas as pd
import numpy as np
import os
import re
from string import punctuation
from collections import Counter
```

```
from keras.models import Sequential
from keras.layers import Dense,Embedding, LSTM
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```
os.chdir("/content/drive/My Drive/Moview review dataset")
data = pd.read_csv('IMDB Dataset.csv')
```

```
data = data.head(3000)
```

```
data.head()
```

```
data.review = data.review.str.lower()
```

```
data.review = data.review.str.replace("<br /><br />","")
```

```
data.review[1]
```

```
print(punctuation)
```

```

data.review = data.review.str.replace('[{}]'.format(punctuation),'')
```

```
reviews_split = data.review
```

```
#print(reviews_split)
```

```
all_text = ' '.join(reviews_split)
words = all_text.split()
```

```
total_words = len(words)
```

```

count_word = Counter(words)
```

```
sorted_words = count_word.most_common(total_words)
```

```
print(count_word)
```

```
print(sorted_words)
```

```
vocab_to_int = {w:i+1 for i,(w,c) in enumerate(sorted_words)}
```

```
data.sentiment = [1 if sentiment == 'positive' else 0 for sentiment in data.sentiment]
```

```
encoded_lables = np.array(data.sentiment)
```

```
tokenizer = Tokenizer(nb_words = len(vocab_to_int),split=' ')
tokenizer.fit_on_texts(data.review.values)
X= tokenizer.texts_to_sequences(data.review.values)
X = pad_sequences(X)
```

```
Y = pd.get_dummies(data['sentiment']).values
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 42)
```

```
len(vocab_to_int)
```

```
embed_dim = 128
lstm_out = 196
batch_size = 32
model = Sequential()
model.add(Embedding(len(vocab_to_int), embed_dim,input_length = X.shape[1], dropout = 0.2))
model.add(LSTM(lstm_out, dropout_U = 0.2, dropout_W = 0.2))
model.add(Dense(2,activation='softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer='adam',metrics = ['accuracy'])
print(model.summary())
```

```
model.fit(X_test, Y_test, batch_size =batch_size, nb_epoch = 3,  verbose = 1)
```
